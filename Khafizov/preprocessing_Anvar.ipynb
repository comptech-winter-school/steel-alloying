Есть такое ощущение, что такие пустые строки надо удалять. Начать именно со строк. Удалим строки, в которых больше 1/3 клеток - пустые.

```python
row_indexes_to_exclude = []
print(f"было {len(df)} строк")

number_of_columns_in_dataframe = len(df.columns)
for index, row in df.iterrows():
    NaN_count = row.isna().sum()
    if NaN_count > number_of_columns_in_dataframe*(1/3):
        row_indexes_to_exclude.append(index)

df = df.drop(df.index[row_indexes_to_exclude])
print(f"осталось {len(df)} строк")
```

Для начала, выберем все колонки, в которых есть хотя бы одна пустая клетка

```python
column_names_with_empty_values = []
print(f"из {len(df.columns)} столбцов НЕ полностью заполнены:")

for column_name in df.columns:
    NaN_count = df[column_name].isna().sum()
    if NaN_count > 0:
        column_names_with_empty_values.append(column_name)
df_empty = df[column_names_with_empty_values]

print(f"{len(df_empty.columns)} столбцов")
```

Узнаем, насколько много данных нам надо заполнять. Пусть, если пустых <1% ячеек, то это будет немного

```python
NaN_counts = []
count_of_rows = len(df_empty)

for column_name in df_empty.columns:
    NaN_count = df_empty[column_name].isna().sum() /count_of_rows * 100
    NaN_counts.append(NaN_count)

_, ax = plt.subplots(1,1, figsize=(20,10), dpi=300)

x_ticks = df_empty.columns
x_pos = [i for i, _ in enumerate(x_ticks)]

trust_level = 1
ax.axhline(y=trust_level, label=f"{trust_level}%")
ax.bar(x_pos, NaN_counts, color='green')
ax.set_xticks(x_pos)
ax.set_ylabel("количество пустых ячеек, %")
ax.legend()
_ = ax.set_xticklabels(x_ticks, rotation='vertical')
```

Уже прорисовались колонки, которые скорее, всего будет легко заполнять и с которыми придется повозиться. Попробуем заодно оценить относительную погрешность т.е. std(X)/mean(X) для каждой колонки X. Пусть, станартная ошибка менее 25% будет приемлимой

```python
def count_var_coef_without_NaN(values):
    mask = np.isnan(values)
    
    if np.sum(mask) == len(values):
        return 0, 0

    NaN_count = np.sum(mask)
    var_coef =  lambda x: np.std(x) / np.mean(x) * 100
    existing_not_NaN_values = values[~mask]

    return var_coef(existing_not_NaN_values), NaN_count


def plot_heat_map_for_nan(df, df_empty):
    steel_grades = df["МАРКА"].unique()
    alloying_factors = [column_name for column_name in df_empty.columns if df_empty[column_name].isna().any()]
    var_coefs = []
    NaN_counts = []
    total_row_counts = []

    grouped = df.groupby("МАРКА")
    for steel_grade in steel_grades:
        var_coef_for_steel_grade = []
        NaN_count_for_steel_grade = []
        total_row_count_for_steel_grade = []

        for alloying_factor in alloying_factors:
            values = grouped.get_group(steel_grade)[alloying_factor].to_numpy()
            vc, NaN_count = count_var_coef_without_NaN(values)
            NaN_count_for_steel_grade.append(NaN_count)
            var_coef_for_steel_grade.append(vc)
            total_row_count_for_steel_grade.append(len(values))

        NaN_counts.append(NaN_count_for_steel_grade)
        var_coefs.append(var_coef_for_steel_grade)
        total_row_counts.append(total_row_count_for_steel_grade)

    var_coefs, NaN_counts, total_row_counts = np.transpose(var_coefs), \
        np.transpose(NaN_counts), np.transpose(total_row_counts)

    fig, ax = plt.subplots(figsize=(15,15), dpi=300)
    im = ax.imshow(var_coefs, cmap="Paired")

    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    fig.colorbar(im, cax=cax, orientation='vertical')

    ax.set_xticks(np.arange(len(steel_grades)))
    ax.set_yticks(np.arange(len(alloying_factors)))

    _ = ax.set_xticklabels(steel_grades, rotation='vertical')
    ax.set_yticklabels(alloying_factors)

    for i in range(len(alloying_factors)):
        for j in range(len(steel_grades)):
            bbox_color, bbox_alpha = ('white', 1) if NaN_counts[i, j] > 0 else ("gray", 0.5)

            if NaN_counts[i, j] / total_row_counts[i, j] > (2/3)*100:
                bbox_color = 'red'

            text = ax.text(j, i, f"{NaN_counts[i, j]} /\n {total_row_counts[i, j]}",
                        ha="center", va="center", color="k",
                        bbox={'facecolor': bbox_color, 'alpha': bbox_alpha, 'pad': 3})

    ax.set_title("dX/<X>")

    plt.show()

plot_heat_map_for_nan(df, df_empty)
```
