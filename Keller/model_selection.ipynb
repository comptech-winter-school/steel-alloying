```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('clean_data.csv', index_col=0)
df.head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
```

```python
X = df.drop([col for col in df.columns if col.startswith('химшлак последний')], axis=1)
X
```

```python
X.info()
```

```python
X = pd.get_dummies(X)
X.info()
```

```python
from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
X = scaler_X.fit_transform(X)
X.shape
```

```python
y = df['химшлак последний R']
y
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso

linreg = Ridge()
linreg.fit(train_X, train_y)
```

```python
linreg.get_params().keys()
```

```python
error = linreg.predict(test_X).sum() / test_y.sum() - 1
print("Percentage error: {:.2f}%".format(error*100))
```

```python
cross_val_score(linreg, X, y, scoring='neg_mean_squared_error').mean()
```

```python
gridCV = GridSearchCV(linreg, {'alpha':np.linspace(0.001, 10, 20)}, cv=5)
gridCV.fit(X, y)
```

```python
gridCV.best_params_, gridCV.best_score_
```

```python
cross_val_score(gridCV.best_estimator_, X, y, scoring='neg_mean_squared_error').mean()
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=10).fit(train_X, train_y)
rf_predict = rf.predict(test_X)
```
